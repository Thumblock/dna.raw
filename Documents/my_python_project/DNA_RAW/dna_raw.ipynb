```python
# Importerar bibliotek för att kunna rita grafer.

import matplotlib.pyplot as plt
```

```python
# Namnger filen som innehåller DNA sekvensen.

filename = "dna_raw.txt"

# Öppnar filen för att läsa in alla rader.

with open(filename, "r") as file:
    lines = file.readlines()
```

```python
# Detta skapar en tom dictionary som används för att lagra sekvenserna.

sequences = {}
current_seq_id = ""

# Loopar igenom varje rad i filen.

for line in lines:
    line = line.strip()  # Tar bort onödiga mellanslag.

    if line.startswith(">"):  # Det är en sekvens id om raden börjar med ">".
        current_seq_id = line  # Sekvens ID sparas.
        sequences[current_seq_id] = ""  # Skapar en tom sekvens för denna ID
    else:
        sequences[current_seq_id] += line.upper()  # Gör dom till stor bokstav.
```

```python
# Dictonary, DNA baser frekvensen lagras.

sequence_counts = {}

# Loopar varje sekvens för att räkna ut baserna.

for seq_id, sequence in sequences.items():
    letter_count = {"A": 0, "T": 0, "C": 0, "G": 0} # Dictionary skapas för att räkna baserna.

    for letter in sequence: # Loopar varje bokstav i sekvensen.
        if letter in letter_count: # Kontrollerar bokstaven.
            letter_count[letter] += 1 # ökar ett steg för aktuell bas.

    sequence_counts[seq_id] = letter_count # Sparar sekvens resultat.

sequence_counts
```

```python
# Skapar en graf för varje sekvens.

for seq_id, letter_count in sequence_counts.items():
    plt.figure(figsize=(6, 4))  # skapar en figur med storleken 6x4
    plt.bar( # Ritar en stapeldiagram
        letter_count.keys(),
        letter_count.values(),
        color=["blue", "red", "green", "orange"],
    ) 
    plt.title(f"Frequency of DNA bases in {seq_id}")
    plt.xlabel("DNA Base") # X-axel för A,T,C,G
    plt.ylabel("Count") # Antal i Y-axeln
    plt.show()
```
